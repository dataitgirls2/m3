```python
from geopy import distance

def get_distance(coor_univ,coor_store):
    dist = distance.distance(coor_univ,coor_store).km
    return dist
```

```python
coor_univ = (37.555273,127.043655) #한양대역
```

```python
get_distance(coor_univ,(37.482716, 126.995413))
```

```python
list_dist = []
for i in df2016_work[['lat','lng']].values:
    coor = i[0],i[1]
    a = get_distance(coor_univ,coor)
    list_dist.append(a)
print(list_dist)
```
