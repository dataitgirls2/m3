```python
import numpy as np
import pandas as pd
import re
```

```python
#배달음식 주문 데이터
df1 = pd.read_csv('CALL_NDELIVERY_06MONTH.csv')
```

```python
df1.head(2)
```

```python
df2 = pd.read_csv('CALL_NDELIVERY_07MONTH.csv')
df3 = pd.read_csv('CALL_NDELIVERY_08MONTH.csv')
df4 = pd.read_csv('CALL_NDELIVERY_09MONTH.csv')
```

```python
#6월부터 9월까지 배달 데이터 합치기
frames = [df1, df2, df3, df4]
call = pd.concat(frames)
```

```python
#6월부터 9월 배달 데이터 출력
call
```

```python
call_food = call[['일자', '시군구', '업종', '통화건수']]
```

```python
call_food.head()
```

```python
food = call_food.groupby(['일자', '시군구', '업종'])[('통화건수')].sum()
temp_food= pd.DataFrame(food)
temp_food
```

```python
weather_df = pd.read_excel('2017summer.xlsx')
```

```python
weather_df.head(2)
```

```python
weather_df['시군구'] = weather_df['지역'].str.replace(r'(강남|강동|강북|강서|관악|광진|구로|금천|남산|노원|도봉|동대문|동작|마포|서대문|서초|성동|성북|송파|양천|영등포|용산|은평|종로|중랑)', 
                         r'\1구')
weather_df.head(25)
```

```python
weather_df['일자'] = weather_df['관측일자']
```

```python
weather_df2 = weather_df[['일자', '시군구', '평균기온', '최저기온', '최대기온', '평균습도', '강수량']]
weather_df2.head()
```

```python
result1 = pd.merge(temp_food, weather_df2, on=['일자', '시군구'])
```

```python
result1 
```

# '업종' 카테고리 추가하기 

```python
result = pd.merge(weather_df2.reset_index(), temp_food.reset_index(),
         on=['일자','시군구'], how='inner').set_index(['일자','시군구'],['업종'])
```
