```python
import numpy as np
import pandas as pd
```

```python
#배달음식 주문 데이터
df1 = pd.read_csv('CALL_NDELIVERY_06MONTH.csv')
```

```python
df1
```

```python
df2 = pd.read_csv('CALL_NDELIVERY_07MONTH.csv')
df3 = pd.read_csv('CALL_NDELIVERY_08MONTH.csv')
```

```python
#7월부터 9월까지 배달 데이터 합치기
frames = [df1, df2, df3]
call = pd.concat(frames)
```

```python
#7월부터 9월 배달 데이터 출력 >_<꺆 40만건
call
```

```python
call['시군구'] = call['시군구'].astype(str)
```

```python
import re
```

```python
#중구 어떻게 잡아요 님들 ㅠㅠ 흑흑 
call['지역'] = call['시군구'].str.extract(r'(\w+)구')
call.head(20)
```

```python
call_food = call[['일자', '업종', '지역', '통화건수']]
```

```python
food= call_food.groupby(['일자', '지역', '업종'])[('통화건수')].sum()
```

```python
temp_food= pd.DataFrame(food)
temp_food
```
