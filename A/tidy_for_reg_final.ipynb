```python
def get_month_from_b_date(df_column):
    b_date = []
    for data in df_column:
        if data[5] == '0':
            month = data[6]
        else:
            month = data[5:7]
        single_date = int(month)
        b_date.append(single_date)

    return pd.Series(b_date)
```

```python
df1['방영월'] = get_month_from_b_date(df1['방영일'])
df1.head()
```

```python
df1 = df1.drop(['방영일'], axis=1)
df1.tail()
```

```python
df1.to_csv('raw_data/list2017.csv', index=False)
```

#### 기존 데이터와 합치기

```python
tidy = pd.read_csv('raw_data/total_tidy.csv')
```

```python
lst.info()
```

```python
tidy.info()
```

```python
reg = pd.DataFrame(tidy.groupby(['식당', '방송'])['방영전', '방영후'].sum()).reset_index()
```

```python
reg['증가'] = reg['방영후'] - reg['방영전']
reg.info()
```

lst와 reg의 데이터 개수가 맞지 않음

```python
lst = lst[lst.식당.isin(reg.식당.values)].reset_index().drop(['index'], axis=1)
lst.info()
```

```python
df = pd.merge(reg, lst, on='식당')
df.info()
```

합쳤더니 df 데이터 개수가 늘어남

아무래도 방송별로 중복된 식당이 있는 것 같음

```python
df.groupby('식당')['방영월'].value_counts().loc[lambda x: x>1].reset_index(name='count')
```

```python
tidy.loc[tidy['식당']=='미타우동']
```

미타우동이 수요미식회에 방영됐을 때는, 이미 맛있는녀석들 방영의 영향력을 받았을 것이기 때문에 데이터를 삭제하기로 함
