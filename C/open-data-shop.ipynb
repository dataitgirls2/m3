```python
import pandas as pd
import folium
```

```python
df_shop = pd.read_csv('data/shop_201806_01.csv', encoding='cp949')
```

```python
df_shop.shape
```

```python
df_shop.head()
```

```python
df_mcdonalds = df_shop.loc[df_shop['상호명'].str.contains('맥도날드') | df_shop['지점명'].str.contains('맥도날드')].copy()
```

```python
df_mcdonalds['DriveThru'] = 'nondt'
```

```python
df_mcdonalds.loc[df_mcdonalds['상호명'].str.contains('DT') | df_mcdonalds['지점명'].str.contains('DT'), 'DriveThru'] = 'DT'
```

```python
geo_df = df_mcdonalds.copy()
```

```python
geo_df.head()
```

```python
# 지도를 초기화 해줄 때 어디를 중심으로 보여줄지 설정합니다.
# 우리가 가져온 데이터프레임 안에 있는 데이터를 기준으로 출력할 수 있도록 위경도의 평균값을 구해옵니다.
map = folium.Map(location=[geo_df['위도'].mean(), geo_df['경도'].mean()], zoom_start=12)
```

```python
for n in geo_df.index:
    # 팝업에 들어갈 텍스트를 지정해 줍니다.
    popup_name = geo_df['DriveThru'][n] + ' - ' + geo_df['도로명주소'][n]
    # 브랜드명에 따라 아이콘 색상을 달리해서 찍어줍니다.
    if geo_df['DriveThru'][n] == 'DT' :
        icon_color = 'green'
    else:
        icon_color = 'blue'    
    
    folium.Marker([geo_df['위도'][n], geo_df['경도'][n]], 
                  popup=popup_name, 
                  icon=folium.Icon(color=icon_color)).add_to(map)
```
